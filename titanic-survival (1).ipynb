*I make a function which extract terms like mr,mrs,dr,others on the basis of this category we plot a barplot we store this category in a new column called title

```python
def categorize_title(name):
    if 'Mr.' in name:
        return 'Mr'
    elif 'Mrs.' in name:
        return 'Mrs'
    elif 'Dr.' in name:
        return 'Dr'
    else:
        return 'Other'

# Create a new 'title' column by applying the categorize_title function
train['title'] = train['Name'].apply(categorize_title)
test['title'] = test['Name'].apply(categorize_title)
# Calculate the survival percentage for each title group within each class
title_survival = train.groupby('title')['Survived'].mean() * 100
sns.barplot(x=title_survival.index,y=title_survival.values)
plt.xlabel('name')
plt.ylabel('Survival Percentage')
plt.title('Survival Percentage by title')
plt.show()
```

**Conclusion**
Ladies and olders have chance of surviving than others

*I dont get a good result so I categorised that family column into 3 subcategorical columns by making a function*

```python

def family_size_cal(no):
    if no==1:
        return'alone'
    elif no>1 and no<5:
        return 'medium'
    else:
        return "large"
  
```

```python
  train['Family_size']=train['Family'].apply(family_size_cal)
test['Family_size']=test['Family'].apply(family_size_cal)
```

```python
train.drop(columns=['SibSp','Parch','Family','Name','Ticket'],inplace=True)
test.drop(columns=['SibSp','Parch','Family','Name','Ticket'],inplace=True)
```

**Important Step**
now this columns are categorical columns we have to convert them into numerical column by get_dummies function ,we also drop the first category to overcome the overfitting ,because lets for embarked its like
C Q S
0 0 1
1 0 0
1 0 0
in this way your model learn that in each row the sum is 1 which trained the model wrong way and gives it a pattern which is known as overfitting,so to avoid this do this

```python
train=pd.get_dummies(train,columns=['Pclass','Sex','Embarked','title','Family_size'],drop_first=True,dtype=np.int64)
test=pd.get_dummies(test,columns=['Pclass','Sex','Embarked','title','Family_size'],drop_first=True,dtype=np.int64)
```

```python
train.head()
```

> Model Training

```python
X=train.iloc[:,1:]
y=train.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

> Accuracy Calculation

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
Xf=test.iloc[:,:]
```

```python
y_final=classifier.predict(Xf)
```
